```python
import os
import torch
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
plt.style.use("ggplot")
```

```python
num_models = 3
path_results = "/develop/results/chip_classifier/many_exps/20230615_EUTKS_NewSim_NewDumbAMA_on_training_data"
```

```python
def load_pkl(path):
    
    data_file = open(path, "rb")
    
    data = pickle.load(data_file)
    
    data_file.close()
    
    return data

def load_csv(path):
    
    return pd.read_csv(path)

def gather_results(path):
    
    results = {}
    
    print("\nGathering Results")
    print("\n--------------------------------------\n")
    
    for current_fold in os.listdir(path):
        
        results[current_fold] = {}
        
        path_fold = os.path.join(path, current_fold)
        
        print("Current Fold: %s" % path_fold)
        
        for current_model in os.listdir(path_fold):
            
            if(current_model == "slurm"):
                continue
            
            path_model = os.path.join(path_fold, current_model)
            
            train_file = os.path.join(path_model, "train_loss.csv")
            valid_file = os.path.join(path_model, "valid_loss.csv")
            test_file = os.path.join(path_model, "test_preds.pkl")
            
            if(os.path.exists(test_file)):
                
                results[current_fold][current_model] = {} 
                results[current_fold][current_model]["train_loss"] = load_csv(train_file)
                results[current_fold][current_model]["valid_loss"] = load_csv(valid_file)
                results[current_fold][current_model]["testing"] = load_pkl(test_file)
                
                print("Found Train Loss: %s" % train_file)
                print("Found Valid Loss: %s" % valid_file)
                print("Found Testing Preds: %s" % test_file)
                
        print("\n--------------------------------------\n")

    return results
```

```python
all_results = gather_results(path_results)
```

```python
def assign_model_names(all_results):

    for fold_key in all_results.keys():

        fold_results = all_results[fold_key]
        
        for i, model_key in enumerate(fold_results.keys()):
            
            if(i == 0):
                name = "ResNet"
            elif(i == 1):
                name = "EfficientNet"
            elif(i == 2):
                name = "AlexNet"
            else:
                raise NotImplementedError

            all_results[fold_key][model_key]["name"] = name
    
    return all_results
```

```python
all_results = assign_model_names(all_results)
```

```python
def score_testing(all_preds, all_truths, name):
    
    new_preds, new_labels = [], []
    for pred_group, label_group in zip(all_preds, all_truths):
        for pred, label in zip(pred_group.numpy(), label_group.numpy()):
            new_preds.append(np.argmax(pred))
            new_labels.append(label)
    
    results = {"name": name,
               
               "f1": f1_score(new_labels, new_preds, average = "macro", zero_division = 0),
               "recall": recall_score(new_labels, new_preds, average = "macro", zero_division = 0),
               "precision": precision_score(new_labels, new_preds, average = "macro", zero_division = 0),

               "class_f1": f1_score(new_labels, new_preds, average = None, zero_division = 0),
               "class_recall": recall_score(new_labels, new_preds, average = None, zero_division = 0),
               "class_precision": precision_score(new_labels, new_preds, average = None, zero_division = 0),

               "accuracy": accuracy_score(new_labels, new_preds),
               "confusion_matrix": confusion_matrix(new_labels, new_preds)}

    return results
    
def organize_results(all_results, num_models):
    
    loss_results = {i: {"train": {"all": []}, "valid": {"all": []}} 
                    for i in range(num_models)}
    
    pred_results = {i: [] for i in range(num_models)}
    
    for fold_key in all_results.keys():
        
        fold_results = all_results[fold_key]
        
        for i, model_key in enumerate(fold_results.keys()):
            
            model_results = fold_results[model_key]
            name = model_results["name"]
            
            loss_results[i]["train"]["name"] = name
            loss_results[i]["valid"]["name"] = name
            loss_results[i]["train"]["all"].append(model_results["train_loss"]["total"])
            loss_results[i]["valid"]["all"].append(model_results["valid_loss"]["total"])
    
            preds, truths = model_results["testing"]["preds"], model_results["testing"]["truths"]
            
            pred_results[i].append(score_testing(preds, truths, name))
    
    for model_key in loss_results:
        
        model_results = loss_results[model_key]
        
        for current_key in model_results.keys():
            
            max_size = max([len(group) for group in model_results[current_key]["all"]])
            
            indices = [i for i, group in enumerate(model_results[current_key]["all"]) if len(group) != max_size]
            
            for index in indices:
                model_results[current_key]["all"].pop(index)
            
            model_results[current_key]["all"] = np.vstack(model_results[current_key]["all"])
            model_results[current_key]["max"] = np.max(model_results[current_key]["all"], axis = 0)
            model_results[current_key]["min"] = np.min(model_results[current_key]["all"], axis = 0)
            model_results[current_key]["avg"] = np.mean(model_results[current_key]["all"], axis = 0)
        
    return loss_results
```

```python
exp_results = organize_results(all_results, num_models)
```

```python
def display_loss(exp_results, num_models, target):
    
    fig, ax = plt.subplots(num_models, 2, figsize = (10, 8))
    
    for i, model_key in enumerate(exp_results.keys()):
        model_results = exp_results[model_key][target]
        for j, current_fold in enumerate(model_results["all"]):
            ax[i, 0].plot(np.arange(0, len(current_fold), 1), current_fold, label = "Fold %s" % j)
            
        ax[i, 0].set_title("%s Results (Folds) - %s" % (target.capitalize(), model_results["name"]))
        ax[i, 0].legend()
    
        min_agg = model_results["min"]
        ax[i, 1].plot(np.arange(0, len(min_agg), 1), min_agg, label = "Min")
        
        max_agg = model_results["max"]
        ax[i, 1].plot(np.arange(0, len(max_agg), 1), max_agg, label = "Max")
        
        avg_agg = model_results["avg"]
        ax[i, 1].plot(np.arange(0, len(avg_agg), 1), avg_agg, label = "Avg")
        
        ax[i, 1].set_title("%s Results (Aggs) - %s" % (target.capitalize(), model_results["name"]))
        
        ax[i, 1].legend()
        
    fig.tight_layout()
```

```python
display_loss(exp_results, num_models, "train")
display_loss(exp_results, num_models, "valid")
```

```python
def display_error(all_results, criteria, all_names, width = 0.08):
    
    # Gather bar heights
    
    targets = [criteria + "_accuracy", criteria + "_recall", criteria + "_precision", criteria + "_f1"]
    
    all_groups = []
    
    for model_results in all_results:
        
        group = []
        for current_key in model_results.keys():
            if(current_key in targets):    
                group.append(model_results[current_key])
        
        all_groups.append(group)
    
    # Gather bar positions
    
    all_positions = []
    
    for i in range(len(all_groups)):
        
        if(i == 0):
            position = np.arange(len(all_groups[0]))
        else:
            position = [ele + width for ele in all_positions[i - 1]]
                
        all_positions.append(position)
           
    fig, ax = plt.subplots(figsize = (8, 4))

    color_range = np.arange(len(all_groups)) + 1
    colors = plt.cm.turbo(color_range / float(max(color_range)))
    for i, (position, group) in enumerate(zip(all_positions, all_groups)):
        
        ax.bar(position, group, width = width, edgecolor = "black", label = all_names[i], color = colors[i])

    ax.set_xlabel("Models")
    ax.set_ylabel("Performance")

    ax.set_ylim([0, 1])
    ax.set_xticks([r + width * 5 for r in range(len(all_groups[0]))], ["Precision", "Recall", "F1", "Accuracy"])
    
    ax.legend(loc = "upper right")
    
    fig.tight_layout()
```
